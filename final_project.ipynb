```python
import numpy as np
```

```python
# given
v = 0.25
theta = 0.5
b = 1
r_bar = 0.03
pi_bar = 0.02
```

```python
matrix_x = np.array([[1, -b, b], [-v, 1, 0], [0,-1-theta,1]])
```

```python
x_inverse = np.linalg.inv(matrix_x)
x_inverse
```

```python
B = np.array([b*r_bar,0,r_bar-theta*pi_bar])
```

```python
A = np.matmul(x_inverse,B)
A
```

```python
Y0 = np.array([0,0.02,0.05])    
yt_1 = Y0
output = [Y0]
A1 = np.matmul(x_inverse,np.array([[0,0,0],[0,1,0],[0,0,0]]))
for i in range(10):

    if i == 0:
        error = np.array([0, 0, 0.01])
    else:
        error = np.array([0, 0, 0])

    yt = A + np.matmul(A1, yt_1) + np.matmul(x_inverse,error)
    yt_1 = yt
    output.append(yt_1)
```

```python
output
```

```python
y_list = [i[0] for i in output]
```

```python
pi_list = [i[1] for i in output]
norm_pi = [i - 0.02 for i in pi_list]
```

```python
i_list = [i[2] for i in output]
norm_i = [i - 0.05 for i in i_list]
```

### plotting the impulse responses
